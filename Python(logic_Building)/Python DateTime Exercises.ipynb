### Iterating through a range of dates

```python
import datetime
start_date = datetime.date(2021,2,1)
end_date = datetime.date(2024,1,26)
delta = datetime.timedelta(days = 1)
while(start_date<=end_date):
    print(start_date,end="\n")
    start_date += delta
```

```python
from datetime import date
from dateutil.rrule import rrule,DAILY
start_date = date(2022,9,1)
end_date = date(2024,1,26)
for d in rrule(DAILY,dtstart=start_date,until=end_date):
    print(d.strftime("%Y-%m-%d"))
```

```python
import pandas as pd
a = pd.date_range(start="1/1/2021",end="26/1/2024")
for i in a:
    print(i.date())
```

### How to add time onto a DateTime object in Python

```python
import datetime
date_and_time = datetime.datetime(2021,8,22,11,2,5)
print("Original Time:",date_and_time)

time_change = datetime.timedelta(minutes = 75)
new_time = date_and_time + time_change
print("Changed time:",new_time)
```

```python
import datetime
date_and_time = datetime.datetime(2021,8,22,11,2,5)
print("Original Time:",date_and_time)

time_change = datetime.timedelta(minutes=2,seconds=10)
new_time = date_and_time + time_change
print("Changed time:",new_time)
```

```python
import datetime
date_and_time = datetime.datetime(2021,8,22,11,2,5)
print("Original Time:",date_and_time)

time_change = datetime.timedelta(hours=35)
new_time = date_and_time + time_change
print("Changed time:",new_time)
```

### How to add timestamp to excel file in Python

```python
import datetime
from openpyxl import Workbook
import time
wb = Workbook()
ws = wb.active
ws.cell(row=1,column=1).value = "Current Date and Time"
ws.cell(row=2,column=1).value = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
time.sleep(2)
ws.cell(row=3, column=1).value = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S') 
time.sleep(2) 
  
# Cell A4 containing the Current Date and Time 
ws.cell(row=4, column=1).value = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S') 
  
# Save the workbook with a 
# filename and close the object 
wb.save('gfg.xlsx') 
wb.close()
```

### Convert string to datetime in Python with timezone

```python
import datetime
date_string = '2021-09-01 15:27:05.004573 +0530'
print("string Datetime:",date_string)
print("datestring class is :",type(date_string))
datetime_obj = datetime.datetime.strptime(date_string,'%Y-%m-%d %H:%M:%S.%f %z')
print("Converted to datetime:",datetime_obj)
print("datetime_obj class is:",type(datetime_obj))
```

### Python datetime to integer timestamp

```python
from datetime import datetime
cur_time = datetime.now()
print("Current Datetime",cur_time)
timestamp = int(round(cur_time.timestamp()))
print("Inteager timestamp of current datetime:",timestamp)
```

```python
from datetime import datetime
dtime = datetime(2018,1,1,20)
print("Datetime:",dtime)
dtimestamp = dtime.timestamp()
print("Inteager timestamp in second:",int(round(dtimestamp)))
milliseconds = int(round(dtimestamp*1000))
print("Inteager timestamp in milisecond:",milliseconds)
```

```python
import datetime
import calendar
d = datetime.datetime(1970,1,1,2,1,0)
ttuple = d.timetuple()
ittimestamp = calendar.timegm(ttuple)
print("TimeStamp in inteager since epoch:",ittimestamp)
```

```python
import datetime
import pytz
dtime = datetime.datetime.now()
timezone = pytz.timezone("Asia/Kolkata")
dtzone = timezone.localize(dtime)
print("Time Zone: ", dtzone.tzinfo)
print("Datetime: ", dtzone)
tstamp = dtzone.timestamp()
print("Inteager timestamp:",int(round(tstamp)))
```

### How to convert a Python datetime.datetime to excel serial date number

```python
import datetime
cur_dt = datetime.datetime.now()
print(cur_dt.strftime("%x %X"))
```

```python
from datetime import date
def convert_date_to_excel_ordinal(day,month,year):
    offset = 693594
    current = date(year,month,day)
    n = current.toordinal()
    return (n - offset)
print(convert_date_to_excel_ordinal(26,1,2024))
```

```python
from datetime import datetime

def excel_date(date1):
    temp = datetime(1899, 12, 30)
    delta = date1 - temp
    return float(delta.days) + (float(delta.seconds) / 86400)

print(excel_date(datetime(2021, 2, 4)))
```

### How to create filename containing date or time in Python

```python
from datetime import datetime
cur_dt = datetime.now().strftime("%Y-%m-%d %H-%M-%S")
print("Current date and time:",cur_dt)
str_cur_dt = str(cur_dt)
file_name = str_cur_dt + ".txt"
file = open(file_name,"w")
print("File Created:",file.name)
file.close()
```

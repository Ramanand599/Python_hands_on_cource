# Matplotlib 
Matplotlib is a plotting library for the Python programming language and its numerical mathmetics extension NumPy.Itprovide an object-Oriented API for embedding plot into application using general-purpose GUI toolkits like Tkinter,wxPython,Qt,or GTK+,

Some of the mejor pros of Matplotlib are:
    - Generally easy to get started for simple plots
    - Support for custom label and texts
    - Great control of every element in figure
    - high-quality output in many formats
    - Very Custmizable in genreal

```python
# compute the x and y coordinate for points on a sin curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
```

```python
# compute the x and y coorditate for points on sin and cosin curve
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# set up subplot grid that has hight 2 and width 1
#and set the first such subplot as active 
plt.subplot(2,1,1)

#Make a first plot
plt.plot(x,y_sin,'r')
plt.title("sin\n")

#set the second subplot as active ,and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,"b")
plt.title("cosin")
plt.show()
```

```python
# Bar plot
x1=[2,8,10]
y1=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x1,y1)
plt.bar(x2,y2,color="g")
plt.title("Bar graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
# Histogram
a=np.array([22,87,5,43,85,97,65,12,4,9,58,44,12,51,7,6,39,30])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

```python
data=[np.random.normal(0,std,100)for std in range(1,4)]
# ractangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```
